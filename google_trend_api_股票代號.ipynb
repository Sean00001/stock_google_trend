```python
import pandas as pd
df = pd.read_csv('stock股票代號.csv')
final=pd.DataFrame()
time1 = '2020-10-26 2021-6-26'
time2 = '2021-6-27 2022-3-17'
df.head(150)
```

```python
#創建csv
colnames = ['日期','所有類別熱度','股票名稱']
header = pd.DataFrame([colnames])
header
header.to_csv('股票代號所有類別.csv',mode ='a', encoding = "Big5",header = False,index=False)
```

```python

from pytrends.request import TrendReq
from pprint import pprint
import time
aa = pd.DataFrame()

cat=0
for i in range(len(df)):
    final = pd.DataFrame()
    out = pd.DataFrame()   
    pytrend = TrendReq(hl='en-US', tz=360)
    keywords =df[i:i+1]['股票代號']
    zx = int(keywords)
    keywords = [str(zx)]
    pytrend.build_payload(
         kw_list = keywords,
         cat=cat,#cat=7 為財經 cat=0 為所有類別
         timeframe= time1,
         geo='TW',
         gprop='')

    out = pytrend.interest_over_time()

    if len(out)== 0:
        print(df[i:i+1]['股票代號'])
        continue;
    else:
        if all(out.isPartial == False):
            del out['isPartial']
#         out.rename(columns={df[i:i+1]['股票名稱'].tolist()[0]: '熱度'}, inplace=True)
        out['股票代號'] = df[i:i+1]['股票代號'].tolist()[0]
        out.to_csv('股票代號所有類別.csv',mode ='a', encoding = "Big5",header = False)
        time.sleep(2)
        final = pd.DataFrame()
        out = pd.DataFrame()
        pytrend.build_payload(
         kw_list = keywords,
         cat=cat,#cat=7 為財經 cat=0 為所有類別
         timeframe= time2,
         geo='TW',
         gprop='')
        out = pytrend.interest_over_time()
        if all(out.isPartial == False):
            del out['isPartial']
#         print(out)
#         out.rename(columns={df[i:i+1]['股票名稱'].tolist()[0]: '熱度'}, inplace=True)
        out['股票代號'] = df[i:i+1]['股票代號'].tolist()[0]
        out.to_csv('股票代號所有類別.csv',mode ='a', encoding = "Big5",header = False)
        time.sleep(1)
```
