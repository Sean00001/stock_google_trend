```python

from pytrends.request import TrendReq
from pprint import pprint
import time
aa = pd.DataFrame()

cat=0
for i in range(len(df)):
    final = pd.DataFrame()
    out = pd.DataFrame()   
    pytrend = TrendReq(hl='en-US', tz=360)
    keywords =df[i:i+1]['股票名稱'].tolist()
    pytrend.build_payload(
         kw_list = keywords,
         cat=cat,#cat=7 為財經 cat=0 為所有類別
         timeframe= time1,
         geo='TW',
         gprop='')

    out = pytrend.interest_over_time()

    if len(out)== 0:
        print(df[i:i+1]['股票名稱'].tolist()[0])
        continue;
    else:
        if all(out.isPartial == False):
            del out['isPartial']
#             print(out)
#         out.rename(columns={df[i:i+1]['股票名稱'].tolist()[0]: '熱度'}, inplace=True)
        out['股票名稱'] = df[i:i+1]['股票名稱'].tolist()[0]
        out.to_csv('股票名稱所有類別.csv',mode ='a', encoding = "Big5",header = False)
        time.sleep(2)
        final = pd.DataFrame()
        out = pd.DataFrame()
        pytrend.build_payload(
         kw_list = keywords,
         cat=cat,#cat=7 為財經 cat=0 為所有類別
         timeframe= time2,
         geo='TW',
         gprop='')
        out = pytrend.interest_over_time()
        if all(out.isPartial == False):
            del out['isPartial']
#             print(out)
#         print(out)
#         out.rename(columns={df[i:i+1]['股票名稱'].tolist()[0]: '熱度'}, inplace=True)
        out['股票名稱'] = df[i:i+1]['股票名稱'].tolist()[0]
        out.to_csv('股票名稱所有類別.csv',mode ='a', encoding = "Big5",header = False)
        time.sleep(1)
```
